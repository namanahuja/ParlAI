```python
import os
from pathlib import Path
import sys
import _pickle as pickle
import json
import copy 
import pandas,csv
import random
```

```python
oneData = []
twoData = []
threeData = []

for root, dirs, files in os.walk("./newData"):
    for file in files:
        if file.endswith(".json"):
            fileName = os.path.join(root, file)
            
            with open(fileName) as json_file:
                
                try:
                    data = json.load(json_file)
                    
                    if 'dataOne' in fileName:
                        oneData.extend(data)
                    
                    elif 'dataTwo' in fileName:
                        twoData.extend(data)
                        
                    else:
                        threeData.extend(data)
                        
                    
                        
                    
                    
                    
                except Exception as e: print(e)
                    
print(oneData[0].keys())
```

```python
allData = []

uniqueContexts = set()

personOne = "<b>Person 1: </b>"
personTwo = "<b>Person 2: </b>"

for root, dirs, files in os.walk("./complete/SIGIR"):
    for file in files:
        if file.endswith(".json"):
            fileName = os.path.join(root, file)
            
            with open(fileName) as json_file:
                
                try:
                    data = json.load(json_file)
                    
                    for conv in data:
                    
                        dialogs = conv['dialogs']
                        
                        for dialog in dialogs:
                            context = dialog['context']
                            
                            
                                
                            
                            contextLen = context['contextLen']
                            fileIndex = context['fileIndex']
                            
                            uniqueStr = "len=" + str(contextLen) + "&" + "index=" + str(fileIndex)
                            
                            if(uniqueStr not in uniqueContexts):
                                uniqueContexts.add(uniqueStr)
                                
                                if(contextLen == 1):
                                    
                                    foundObj = [x for x in oneData if 
                                                   x['index'] == fileIndex][0]
                                    
                                   
                                    conversation = foundObj['conversation']
                                    
                                    
                                    
                                    suggestions = foundObj['suggestions'][:-1]
                                    
                                    
                                        
                                    for i,sugg in enumerate(suggestions):
                                        if(context['userResponse'].strip() == sugg):
                                            continue
                                        
                                        suggObj = {}
                                        
                                        suggObj['suggIndex'] = i
                                        currStr = ""
                                        leftFloat = True

                                        for j in range(0, len(conversation)):
                                            if (leftFloat is True):
                                                currStr += "<div class='conv leftConv'>" + personOne + conversation[j] + "</div>"
                                            else:
                                                currStr += "<div class='conv rightConv'>" + personTwo + conversation[j] + "</div>"

                                            leftFloat = not leftFloat


                                        if (leftFloat is True):
                                            respStr = "<div class='response leftConv'>" + personOne + sugg + "</div>"
                                        else:
                                            respStr = "<div class='response rightConv'>" + personTwo + sugg + "</div>"
    
                                        
                                        suggObj['contextLen'] = contextLen
                                        suggObj['fileIndex'] = fileIndex
                                        suggObj['conversation'] = currStr + "<br />" +respStr
                                        
                                        allData.append(suggObj)
                            
                                
                                   
                                
                                if(contextLen == 2):
                                    
                                    
                                    foundObj = [x for x in twoData if 
                                                   x['index'] == fileIndex][0]
                                    
                                   
                                    conversation = foundObj['conversation']
                                    
                                    
                                    
                                    suggestions = foundObj['suggestions'][:-1]
                                    
                                    for i,sugg in enumerate(suggestions):
                                        if(context['userResponse'].strip() == sugg):
                                            continue
                                        
                                        suggObj = {}
                                        
                                        suggObj['suggIndex'] = i
                                        currStr = ""
                                        leftFloat = True

                                        for j in range(0, len(conversation)):
                                            if (leftFloat is True):
                                                currStr += "<div class='conv leftConv'>" + personOne + conversation[j] + "</div>"
                                            else:
                                                currStr += "<div class='conv rightConv'>" + personTwo + conversation[j] + "</div>"

                                            leftFloat = not leftFloat


                                        if (leftFloat is True):
                                            respStr = "<div class='response leftConv'>" + personOne + sugg + "</div>"
                                        else:
                                            respStr = "<div class='response rightConv'>" + personTwo + sugg + "</div>"
    
                                        
                                        suggObj['contextLen'] = contextLen
                                        suggObj['fileIndex'] = fileIndex
                                        suggObj['conversation'] = currStr + "<br />" +respStr
                                        
                                        allData.append(suggObj)
                            
                                
         
                                
                                if(contextLen == 3):

                                    foundObj = [x for x in threeData if 
                                                   x['index'] == fileIndex][0]
                                    
                                   
                                    conversation = foundObj['conversation']
                                    
                                    
                                    
                                    suggestions = foundObj['suggestions'][:-1]
                                    
                                    for i,sugg in enumerate(suggestions):
                                        if(context['userResponse'].strip() == sugg):
                                            continue
                                        
                                        suggObj = {}
                                        
                                        suggObj['suggIndex'] = i
                                        currStr = ""
                                        leftFloat = True

                                        for j in range(0, len(conversation)):
                                            if (leftFloat is True):
                                                currStr += "<div class='conv leftConv'>" + personOne + conversation[j] + "</div>"
                                            else:
                                                currStr += "<div class='conv rightConv'>" + personTwo + conversation[j] + "</div>"

                                            leftFloat = not leftFloat


                                        if (leftFloat is True):
                                            respStr = "<div class='response leftConv'>" + personOne + sugg + "</div>"
                                        else:
                                            respStr = "<div class='response rightConv'>" + personTwo + sugg + "</div>"
    
                                        
                                        suggObj['contextLen'] = contextLen
                                        suggObj['fileIndex'] = fileIndex
                                        suggObj['conversation'] = currStr + "<br />" +respStr
                                        
                                        allData.append(suggObj)
                            
                                
                                
                except Exception as e: print(e)

random.shuffle(allData)   
print(len(uniqueContexts))
print(len(allData))
```

```python
CSVFile = open('./data_file_SIGIR_suggestions.csv', 'w') 
csv_writer = csv.writer(CSVFile) 

header = allData[0].keys() 
csv_writer.writerow(header) 

for data in allData:
    csv_writer.writerow(data.values()) 
  
CSVFile.close() 
```
