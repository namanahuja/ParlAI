```python
import csv
import json
```

```python
resultFile = 'Batch_273011_batch_results.csv'

with open(resultFile) as f:
    reader = csv.DictReader(f)
    rows = list(reader)

```

```python
rows
```

```python
firstRow = rows[0]
fileNumbers = []
for k in firstRow:
    if('Input.contextLen-vt' in k):
        
        fileNumbers.append(int(k.split("-")[-1]))
        
nFiles = max(fileNumbers)
nFiles
```

```python
rows[0]
```

```python
allData = []
badHITS = []

for row in rows:
    
    
    for i in range(nFiles):
        dataObj = {}
        
        identifier = '-vt-' + str(i+1)
        
        
        for key in row:
            
            if key in ['WorkTimeInSeconds', 'HITId', 'HITTypeId', 'AssignmentId', 'WorkerId']:
                dataObj[key] = row[key]
            
            splits = key.split(identifier)
            
            if(len(splits) > 1):
                keyName = "".join(splits)
                keyVal = row[key]
                if(keyName == 'Input.conversation'):
                    continue
                    
                if(keyName == 'Input.contextLen' and keyVal == -1):
                    # Criteria to reject HITS
                    
                    badHITS.append(row['HITId'])
                    
                
                dataObj[keyName] = keyVal
                  
        
        dataObj['Input.contextLen'] != '-1' and dataObj['Input.contextLen'] != '-2' and allData.append(dataObj)

print(len(allData))
with open('responseJudgement.json', 'w') as f:
    json.dump(allData, f, indent=4)
```

```python
for row in rows:
    
    if row['HITId'] in badHITS:
        row['Reject'] = 'x'
    
    
```

```python
writeFile = 'Edit_' + resultFile

CSVFile = open(writeFile, 'w') 
csv_writer = csv.writer(CSVFile) 

header = rows[0].keys() 
csv_writer.writerow(header) 

for data in rows:
    csv_writer.writerow(data.values()) 
  
CSVFile.close() 
```
