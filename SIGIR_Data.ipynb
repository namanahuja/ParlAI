```python
allData = []
for root, dirs, files in os.walk("./complete/clean"):
    for file in files:
        if file.endswith(".json"):
            fileName = os.path.join(root, file)
            with open(fileName) as json_file:
                cleanObj = {}
                dialogs = []
                
                try:
                    data = json.load(json_file)
                    if(data['suggestionsConfig'] == 'no'):
                        
                        fileName = data['currentFileName']
                        allDialogs = data['dialogs']
                        
                        for i in range(1, len(allDialogs)):
                            context = allDialogs[i]['context']
                            context['fileName'] = fileName
                            allData.append(context)
                    
                except Exception as e: print(e)

randomData = random.choices(allData, k=50)
filePath = './complete/SIGIR/' + 'noSugestions.json'
with open(filePath, 'w') as outfile:
    json.dump(randomData, outfile)
```
